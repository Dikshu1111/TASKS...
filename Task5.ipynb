 1: BAR CHARTS

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Years for comparison
years = [2004, 2007, 2010, 2013, 2016]

# Sample runs data
sachin_runs = [1000, 1200, 1100, 900, 800]
kohli_runs = [0, 500, 1000, 1200, 1300]
sehwag_runs = [800, 950, 1020, 850, 600]

x = np.arange(len(years))
width = 0.25

plt.figure(figsize=(10, 6))
plt.bar(x - width, sachin_runs, width, label='Sachin')
plt.bar(x, kohli_runs, width, label='Kohli')
plt.bar(x + width, sehwag_runs, width, label='Sehwag')

plt.xlabel('Year')
plt.ylabel('Runs Scored')
plt.title('Runs Comparison Over Selected Years')
plt.xticks(x, years)
plt.legend()
plt.tight_layout()
```

```python
# Total runs in debut 5 years
debut_runs = {
    'Sachin': sum([200, 500, 600, 700, 750]),
    'Kohli': sum([300, 650, 700, 800, 900]),
    'Sehwag': sum([150, 400, 550, 600, 500])
}

players = list(debut_runs.keys())
total_runs = list(debut_runs.values())

plt.figure(figsize=(8, 4))
plt.barh(players, total_runs, color=['blue', 'green', 'orange'])
plt.xlabel('Total Runs')
plt.title('Total Runs in First 5 Years')
plt.tight_layout()
```

2: PIE CHART

```python
labels = ['Android (Samsung, Xiaomi, Oppo, Vivo)', 'iOS (iPhone)', 'Others']
sizes = [75, 20, 5]
colors = ['#66b3ff', '#99ff99', '#ffcc99']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Mobile OS Market Share')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.barh(labels, sizes, color=colors)
plt.xlabel('Market Share (%)')
plt.title('Mobile OS Market Share (Bar Chart)')
plt.tight_layout()
plt.show()
```


The **horizontal bar chart** is easier to understand because it clearly shows the market share using **bar lengths**. It's simpler to compare categories like Android and iOS this way, especially when the difference is small.


1) Plot a Line Chart:
   Plot y = 3x + 2 for x in the range 0 to 10. Set the x-axis and y-axis labels,
   and give the plot a title.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 11)
y = 3 * x + 2

plt.figure()
plt.plot(x, y)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Line Plot of y = 3x + 2')
plt.show()
```

3) Bar Chart with Annotations: Draw a vertical bar chart for five categories of your choice.
Annotate the value on top of each bar.

```python

categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 12, 36, 29]

plt.figure()
bars = plt.bar(categories, values, color='skyblue')
for bar in bars:
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height(), str(bar.get_height()), 
             ha='center', va='bottom')
plt.title('Bar Chart with Annotations')
plt.show()

```

4) Multiple Lines in One Plot: Plot three mathematical functions (sin, cos, exp) on the same
axes, each with different colors and line styles. Add a legend.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
plt.figure()
plt.plot(x, np.sin(x), label='sin(x)', linestyle='-', color='blue')
plt.plot(x, np.cos(x), label='cos(x)', linestyle='--', color='green')
plt.plot(x, np.exp(x/10), label='exp(x/10)', linestyle='-.', color='red')
plt.legend()
plt.title('Some Mathematical Functions on One Plot')
plt.show()
```

6) Subplots Grid: Create a 2x2 grid of subplots: line plot, bar chart, scatter plot, and
histogram, each in a different subplot. Add an overall (supertitle) for the figure.

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
x = np.arange(0, 10)
axs[0, 0].plot(x, x**2)
axs[0, 0].set_title('Line Plot')

# Bar chart
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
axs[0, 1].bar(categories, values)
axs[0, 1].set_title('Bar Chart')

# Scatter plot
x_scatter = np.random.rand(20)
y_scatter = np.random.rand(20)
axs[1, 0].scatter(x_scatter, y_scatter, color='red')
axs[1, 0].set_title('Scatter Plot')

# Histogram
data = np.random.randn(500)
axs[1, 1].hist(data, bins=20, color='green')
axs[1, 1].set_title('Histogram')

fig.suptitle('2x2 Subplot Grid')
plt.tight_layout()
plt.show()
```

7) Stack Plot: Plot the cumulative sales of 3 products over 12 months as a stack plot using
made-up data.

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13)
product1 = np.random.randint(20, 50, size=12)
product2 = np.random.randint(30, 60, size=12)
product3 = np.random.randint(10, 40, size=12)

plt.figure()
plt.stackplot(months, product1, product2, product3, labels=['Product A', 'Product B', 'Product C'])
plt.legend(loc='upper left')
plt.title('Cumulative Sales Over 12 Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

8) Boxplot Using DataFrame: Given a DataFrame of student exam scores for three subjects,
visualize the data distribution with a boxplot.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

scores = {
    'Math': np.random.randint(40, 100, 30),
    'Science': np.random.randint(35, 95, 30),
    'English': np.random.randint(45, 90, 30)
}

df = pd.DataFrame(scores)

plt.figure()
df.boxplot()
plt.title('Student Exam Score Distribution')
plt.ylabel('Scores')
plt.show()
```

9) Pie Chart with Explode: Create a pie chart showing the market share of four companies.
Explode the largest segment and display percentage values on the chart.

```python
import matplotlib.pyplot as plt

companies = ['Company A', 'Company B', 'Company C', 'Company D']
market_share = [25, 35, 20, 20]
explode = [0, 0.1, 0, 0]  # Explode the largest segment here (Company B)

plt.figure()
plt.pie(market_share, labels=companies, explode=explode, autopct='%1.1f%%', startangle=140)
plt.title('Market Share of Companies')
plt.show()
```

10) Axis and Legend Customization: For a random line plot, reverse the x-axis, move the
legend outside the plot, and increase legend font size.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure()
plt.plot(x, y, label='sin(x)', color='hotpink')

plt.gca().invert_xaxis()  # Reverse x-axis

plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), fontsize='large')
plt.title('Customized Line Plot')
plt.show()
```
